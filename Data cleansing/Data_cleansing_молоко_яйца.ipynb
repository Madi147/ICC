```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
from pylab import rcParams
from IPython.display import display
import math

import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning
warnings.simplefilter("ignore", category=InterpolationWarning)
pd.set_option('display.max_columns', 130)

# pd.set_option('future.no_silent_downcasting', False)
```

```python
# open and read excel file
x_vars = pd.read_excel("../input data/Общий индекс цен сх м-к-м.xlsx")
x_vars.head(5)
```

```python
x_vars = x_vars.rename(columns={x_vars.columns[0]: "Показатель", x_vars.columns[1]: "Регион"})
x_vars.head(5)
```

```python
x_vars.info()
```

```python
x_vars.columns = list(x_vars.columns[:2]) + list(pd.to_datetime(x_vars.columns[-123:]))
x_vars = x_vars.rename(columns={col: col.strftime("%Y-%m") for col in x_vars.columns[-123:]})
x_vars = x_vars.iloc[:, :-3]
x_vars.head(5)
```

```python
x_vars['Показатель'].unique()
```

```python
x_vars = x_vars[x_vars['Показатель'].isin(['Молоко', 'Яйца'])]
x_vars
```

```python
mapping = {
    "Молоко": "Цена: Молоко",
    "Яйца": "Цена: Яйца"         # already fine
}

x_vars["Показатель"] = x_vars["Показатель"].replace(mapping)
x_vars['Показатель'].unique()
```

```python
x_vars['Регион'].unique()
```

```python
x_vars['Регион'].nunique()
```

```python
x_vars["Регион"] = x_vars["Регион"].str.strip() # strip unnecessary spaces
x_vars = x_vars[x_vars["Регион"] != "ЮЖНО-КАЗАХСТАНСКАЯ ОБЛАСТЬ"] # remove ЮКО
x_vars["Регион"].unique()
```

```python
region_mapping = {
    "Г.АСТАНА": "ГАСТАНА",
    "Г.АЛМАТЫ": "ГАЛМАТЫ",
    "Г.ШЫМКЕНТ": "ГШЫМКЕНТ"          # already fine
}

x_vars["Регион"] = x_vars["Регион"].replace(region_mapping)
x_vars['Регион'].unique()
```

```python
x_vars.sample(20)
```

```python
# Step 1: Identify month columns
month_cols = [col for col in x_vars.columns if col not in ["Показатель", "Регион"]]

# Step 2: Count NaNs row-wise and group by Показатель and Регион
null_counts = (
    x_vars[["Показатель", "Регион"] + month_cols]
    .set_index(["Показатель", "Регион"])[month_cols]
    .isna()
    .sum(axis=1)
    .reset_index(name="Число пропущенных значений")
)

# Step 3: Group total NaNs by Показатель and Регион
null_summary = null_counts.groupby(["Показатель", "Регион"])["Число пропущенных значений"].sum().reset_index()
null_summary[null_summary['Число пропущенных значений'] != 0]
```

```python
rows_to_drop = (
    (x_vars["Показатель"] == "Цена: Яйца") & (x_vars["Регион"] == "КЫЗЫЛОРДИНСКАЯ ОБЛАСТЬ") |
    (x_vars["Показатель"] == "Цена: Яйца") & (x_vars["Регион"] == "МАНГИСТАУСКАЯ ОБЛАСТЬ") |
    (x_vars["Показатель"] == "Цена: Яйца") & (x_vars["Регион"] == "ОБЛАСТЬ АБАЙ") |
    (x_vars["Показатель"] == "Цена: Яйца") & (x_vars["Регион"] == "ОБЛАСТЬ ҰЛЫТАУ") |
    (x_vars["Регион"] == "ГШЫМКЕНТ")
)

# Drop those rows
x_vars = x_vars[~rows_to_drop]
```

```python
# Step 1: Identify month columns
month_cols = [col for col in x_vars.columns if col not in ["Показатель", "Регион"]]

# Step 2: Count NaNs row-wise and group by Показатель and Регион
null_counts = (
    x_vars[["Показатель", "Регион"] + month_cols]
    .set_index(["Показатель", "Регион"])[month_cols]
    .isna()
    .sum(axis=1)
    .reset_index(name="Число пропущенных значений")
)

# Step 3: Group total NaNs by Показатель and Регион
null_summary = null_counts.groupby(["Показатель", "Регион"])["Число пропущенных значений"].sum().reset_index()
null_summary[null_summary['Число пропущенных значений'] != 0]
```

```python
# Replace commas with dots in all cells
x_vars = x_vars.replace(',', '.', regex=True)

# Convert all columns (except identifiers) to numeric
cols_to_convert = [col for col in x_vars.columns if col not in ["Показатель", "Регион"]]
x_vars[cols_to_convert] = x_vars[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
month_cols = [col for col in x_vars.columns if col not in ["Показатель", "Регион"]]
# x_vars[month_cols] = x_vars[month_cols].apply(pd.to_numeric, errors='coerce')
x_vars[month_cols] = x_vars[month_cols].ffill(axis=1).fillna(100.0)

x_vars[month_cols] = x_vars[month_cols] / 100

x_vars[month_cols] = x_vars[month_cols].cumprod(axis=1) * 100  # base = 100

x_vars.head()
```

```python
# Step 1: Define target regions and periods
regions_to_null = ['ОБЛАСТЬ ЖЕТІСУ', 'ОБЛАСТЬ ҰЛЫТАУ', 'ОБЛАСТЬ АБАЙ']
periods_to_null = pd.date_range("2015-01", "2022-05", freq="MS").strftime("%Y-%m")

# Step 2: Replace matching values with NaN
mask = (
    x_vars["Регион"].isin(regions_to_null) &
    x_vars.columns[2:].to_series().isin(periods_to_null).any()
)

# Step 3: Set values to NaN across all columns for those rows and columns
value_columns = x_vars.columns[2:]  # assuming first 3 columns are identifiers

for period in periods_to_null:
    if period in value_columns:
        x_vars.loc[x_vars["Регион"].isin(regions_to_null), period] = np.nan
x_vars.tail(5)
```

```python
# Step 1: Define target regions and periods
regions_to_null2 = ['ГШЫМКЕНТ', 'ТУРКЕСТАНСКАЯ ОБЛАСТЬ']
periods_to_null2 = pd.date_range("2015-01", "2018-05", freq="MS").strftime("%Y-%m")

# Step 2: Replace matching values with NaN
mask = (
    x_vars["Регион"].isin(regions_to_null2) &
    x_vars.columns[2:].to_series().isin(periods_to_null2).any()
)

# Step 3: Set values to NaN across all columns for those rows and columns
value_columns = x_vars.columns[2:]  # assuming first 3 columns are identifiers

for period in periods_to_null2:
    if period in value_columns:
        x_vars.loc[x_vars["Регион"].isin(regions_to_null2), period] = np.nan
x_vars.tail(5)
```
